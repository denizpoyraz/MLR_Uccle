```python
from LOTUS_regression.regression import mzm_regression
from LOTUS_regression.predictors import load_data
import LOTUS_regression.tests as tests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from LOTUS_regression.predictors.seasonal import add_seasonal_components
```

```python
pwlt = load_data('pred_baseline_pwlt.csv')
ilt = load_data('pred_baseline_ilt.csv')
iltq = load_data('pred_baseline_ilt.csv')
pwlt[0:5]
```

```python
pwlt_noqbo = pwlt.drop(['qboA', 'qboB'], axis=1)
ilt_noqbo = ilt.drop(['qboA', 'qboB'], axis=1)
```

```python
s2_omps_osiris = tests.load_example_data('S2_OSIRIS_OMPS_alt_nd_sample.csv')

plt.figure()
s2_omps_osiris['relative_anomaly'].plot()
```

```python
# pwlt
pwlt, s2_omps_osiris = pd.DataFrame.align(pwlt, s2_omps_osiris, axis=0)

# (nsamples, npredictors) matrix
Xpw = pwlt.values

# (nsamples) array of observations
Ypw = s2_omps_osiris['relative_anomaly'].values

regression_output_pw = mzm_regression(Xpw, Ypw)

plt.figure()
plt.plot(s2_omps_osiris.index, Ypw)

plt.plot(s2_omps_osiris.index, regression_output_pw['fit_values'])

```

```python
# pwlt no QBO
pwlt_noqbo, s2_omps_osiris = pd.DataFrame.align(pwlt_noqbo, s2_omps_osiris, axis=0)

# (nsamples, npredictors) matrix
Xpw_nq = pwlt_noqbo.values

# (nsamples) array of observations
Ypw_nq = s2_omps_osiris['relative_anomaly'].values

regression_output_pw_nq = mzm_regression(Xpw_nq, Ypw_nq)

plt.figure()
plt.plot(s2_omps_osiris.index, Ypw_nq)

plt.plot(s2_omps_osiris.index, regression_output_pw_nq['fit_values'])

```

```python
# ilt
ilt, s2_omps_osiris = pd.DataFrame.align(ilt, s2_omps_osiris, axis=0)

# (nsamples, npredictors) matrix
Xi = ilt.values

# (nsamples) array of observations
Yi = s2_omps_osiris['relative_anomaly'].values

regression_output_i = mzm_regression(Xi, Yi)

plt.figure()
plt.plot(s2_omps_osiris.index, Yi)

plt.plot(s2_omps_osiris.index, regression_output_i['fit_values'])

```

```python
# ilt no QBO
ilt_noqbo, s2_omps_osiris = pd.DataFrame.align(ilt_noqbo, s2_omps_osiris, axis=0)

# (nsamples, npredictors) matrix
Xi_nq = ilt_noqbo.values

# (nsamples) array of observations
Yi_nq = s2_omps_osiris['relative_anomaly'].values

regression_output_i_nq = mzm_regression(Xi_nq, Yi_nq)

plt.figure()
plt.plot(s2_omps_osiris.index, Yi_nq)

plt.plot(s2_omps_osiris.index, regression_output_i_nq['fit_values'])

```

```python
# now test Uccle data
uccle =   pd.read_csv('/home/poyraden/MLR_Uccle/Files/1km_monthlymean_all_relative.csv')
uccle['date'] =  pd.to_datetime(uccle['date'], format='%Y-%m')
uccle.set_index('date', inplace=True)
uccle_time = uccle.loc['1977-02-01':'2017-06-01']
```

```python
#ilt normal
ilt, uccle_time = pd.DataFrame.align(ilt, uccle_time, axis=0)

# (nsamples, npredictors) matrix
Xu = ilt.values

# (nsamples) array of observations
Yu = uccle_time['10km_ds'].values

regression_output_u = mzm_regression(Xu, Yu)

plt.figure()
plt.plot(uccle_time.index, Yu)

plt.plot(uccle_time.index, regression_output_u['fit_values'])

```

```python
# now ilt normal residuals
plt.plot(uccle_time.index, regression_output_u['residual'])
```

```python
#ilt with qbo cycle
ilt_with_seasons, uccle_time = pd.DataFrame.align(ilt_with_seasons, uccle_time, axis=0)

# (nsamples, npredictors) matrix
Xq = ilt_with_seasons.values

# (nsamples) array of observations
Yq = uccle_time['10km_ds'].values

regression_output_q = mzm_regression(Xq, Yq)

plt.figure()
plt.plot(uccle_time.index, Yq)

plt.plot(uccle_time.index, regression_output_q['fit_values'])

```

```python
# ilt qith qbo seasonals
plt.plot(uccle_time.index, regression_output_q['residual'])
```

```python
# last check, check the residuals of seasonal data with seasonalized ilt
uccles = pd.read_csv('/home/poyraden//MLR_Uccle/Files/1km_monthlymean.csv')
uccles.rename(columns={'Unnamed: 0':'date'}, inplace=True)
uccles['date'] =  pd.to_datetime(uccles['date'], format='%Y-%m')
uccles.set_index('date', inplace=True)
uccles_time = uccles.loc['1977-02-01':'2017-06-01']
uccles_time = uccles_time[uccles_time['10km']>0]
```

```python
predictors = load_data('pred_baseline_ilt.csv')
predictors_with_seasons = add_seasonal_components(predictors, {'pre_const': 4, 'post_const':4, 'gap_cons':4})
```

```python
predictors_with_seasons, uccles_time = pd.DataFrame.align(predictors_with_seasons, uccles_time, axis=0)

# (nsamples, npredictors) matrix
Xs = predictors_with_seasons.values

# (nsamples) array of observations
Ys = uccles_time['10km'].values

regression_output_s = mzm_regression(Xs, Ys)

plt.figure()
plt.plot(uccles_time.index, Ys)

plt.plot(uccles_time.index, regression_output_s['fit_values'])

```

```python
# not the residuals
plt.plot(uccles_time.index, regression_output_s['residual'])
```
